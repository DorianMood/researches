```python
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
x = torch.linspace(0, 10, steps=100)
y = torch.randint(10, [100])

fig = plt.figure()

ax = fig.add_subplot(111)
ax.scatter(x, y)
```

```python
import scipy.interpolate

t, c, k = scipy.interpolate.splrep(x.numpy(), y.numpy(), k=1)
spline_basis_linear = scipy.interpolate.BSpline(t, c, k, extrapolate=False)

print(k)

t, c, k = scipy.interpolate.splrep(x.numpy(), y.numpy(), k=2)
spline_basis_quadratic = scipy.interpolate.BSpline(t, c, k, extrapolate=False)

# Draw spline

xx = torch.linspace(0, 10)

fig, ax = plt.subplots(1, 2, figsize=[14, 4])

ax[0].plot(xx, spline_basis_linear(xx))
ax[0].scatter(x, y, c='red')
ax[0].set_title('linear')

ax[1].plot(xx, spline_basis_quadratic(xx))
ax[1].scatter(x, y, c='red')
ax[1].set_title('quadratic')
```
